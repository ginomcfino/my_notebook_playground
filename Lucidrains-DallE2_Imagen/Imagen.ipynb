```python
import torch
from imagen_pytorch import Unet, Imagen
```

```python
unet1 = Unet(
    dim = 32,
    cond_dim = 512,
    dim_mults = (1, 2, 4, 8),
    num_resnet_blocks = 3,
    layer_attns = (False, True, True, True),
    layer_cross_attns = (False, True, True, True)
)
```

```python
unet2 = Unet(
    dim = 32,
    cond_dim = 512,
    dim_mults= (1, 2, 4, 8),
    num_resnet_blocks = (2, 4, 8, 8),
    layer_attns = (False, False, False, True),
    layer_cross_attns = (False, False, False, True)
)
```

```python
imagen = Imagen(
    unets = (unet1, unet2),
    image_sizes = (64, 256),
    timesteps = 1000,
    cond_drop_prob = 0.1
)
#.cuda()
```

```python
text_embeds = torch.randn(4, 256, 768)
text_masks = torch.ones(4, 256).bool()
images = torch.randn(4, 3, 256, 256)
#.cuda()
```

```python
for i in (1, 2):
    loss = imagen(images, text_embeds=text_embeds, text_masks = text_masks, unet_number = i)
    loss
```
